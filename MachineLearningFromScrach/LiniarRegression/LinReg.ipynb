```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Salary Data.csv")
```

```python
sns.scatterplot(data=df)
```

```python

def loss_function(x, y, m, b)-> int:
    n = len(y)
    s = 0
    for i in range(n):
        s += (y[i] - (x[i]*m + b))**2

    return s/n
```

```python
def gradient_function(x, y, m, b):
    n = len(y)
    dt_dm, dt_db = 0, 0

    for i in range(n):
        dt_db += ((x[i]*m + b) - y[i])
        dt_dm += x[i] * ((x[i]*m + b) - y[i])

    dt_db *= 2/n
    dt_dm *= 2/n

    return dt_db, dt_dm
```

```python
def lin_reg_algorithm(iterations, alpha):
    b, m = 0, 0
    x, y = np.array(df["YearsExperience"].values), np.array(df["Salary"].values)

    for i in range(iterations):
        cur_alpha = alpha / (1 + 0.1*i)
        vector_b, vector_m = gradient_function(x, y, m ,b)

        b = b - cur_alpha*vector_b
        m = m - cur_alpha*vector_m
    return m, b
```

```python
m, b = lin_reg_algorithm(10000, 0.1)

print(f"Final result: y = {m}x + {b}")
```

```python
x, y = np.array(df["YearsExperience"].values), np.array(df["Salary"])
plt.scatter(x, y, label='Data Points')

x_vals = np.linspace(min(x), max(x), 100)
y_vals = m * x_vals + b
plt.plot(x_vals, y_vals, color='red', label='Regression Line')

plt.xlabel("YearsExperience")
plt.ylabel("Salary")
plt.legend()
plt.show()
```
